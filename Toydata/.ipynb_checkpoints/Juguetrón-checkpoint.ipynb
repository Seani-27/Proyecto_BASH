```python
juguetes=driver.find_elements(By.CLASS_NAME,"dfd-card-title")
#print(len(juguetes))
#print(juguetes[0].text)
```

```python
lista_nombres=[]
for i in range(0,len(juguetes)):
    try:
        lista_nombres.append(juguetes[i].text)
    except:
        lista_nombres.append(np.nan)
#print(lista_nombres)
```

```python
lista_fecha=[]
for i in range (0, len(juguetes)):
    lista_fecha.append(datetime.now().strftime("%d/%m/%Y"))
```

```python
lista_urls=[]
lista_marca=[]
for i in range(0, len(juguetes)): 
    lista_urls.append(driver.find_elements(By.CLASS_NAME,"dfd-card-link")[i].get_attribute("href"))

lista_urls = [i for i in lista_urls if str(i) != "nan"]
#print(len(lista_urls))
(lista_urls)
```

```python
lista_marca=[]
lista_precio=[]
lista_precio_promocion=[]

for j in lista_urls:
    driver.get(j)
    time.sleep(8)
    try:
        lista_marca.append(driver.find_elements(By.CLASS_NAME, "base")[0].text)
    except:
        lista_marca.append(np.nan)
    try:
        lista_precio.append(driver.find_elements(By.CLASS_NAME, "price-wrapper ")[0].text.replace("$","").replace("'","").replace("'",""))
    except:
        lista_precio.append(np.nan)
    try:
        lista_precio_promocion.append(driver.find_elements(By.CLASS_NAME, "old-price")[0].text.replace("$","").replace("'","").replace("'",""))
        
    except:
        lista_precio_promocion.append(np.nan)
#time.sleep(500)
```
